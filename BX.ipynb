```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("data.csv")
```

画出散点图矩阵，查看变量之间的关系

```python
pd.plotting.scatter_matrix(data,figsize=(10,10))
```

求变量间的相关性矩阵，A与B，几乎线性相关

```python
data.corr(method='pearson', min_periods=1)
```

但通过观察关系图，我们相信A与B中有B的二次方项，所以引入一个新变量B2

```python
b2 = []
for i in data['B']:
    b2.append(i*i)
data['B2'] = b2
data
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# 分成训练集与测试集

test = data.sample(frac=0.2, replace=False)
train = data.loc[~data.index.isin(test.index)]

X_train = train[['B','B2']]
Y_train = train['A']
X_test = test[['B','B2']]
Y_test = test['A']

# 创建并拟合模型
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
print (model.intercept_)
print (model.coef_)
```

```python
Y_pict = model.predict(X_test)
```

```python
from sklearn import metrics
# 用scikit-learn计算MSE
print ("MSE:",metrics.mean_squared_error(Y_pict, Y_test))
# 用scikit-learn计算RMSE
print ("RMSE:",np.sqrt(metrics.mean_squared_error(Y_pict, Y_test)))
```

```python
fig, axs = plt.subplots(1, 1)
axs.plot(range(1,201), Y_pict, range(1,201), Y_test)
axs.set_xlabel('we have 200 sample in test set')
axs.set_ylabel('value')
axs.grid(True)

fig.tight_layout()
plt.show()
```
